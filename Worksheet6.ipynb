* If you were to encode the "Student_id" value in one of these channels, why would `"Student_id:N"` make much more sense than `"Student_id:Q"` or `"Student_id:O"`?  Explain in a markdown cell.

Using "Student_id:N" makes more sense since student id's are a nominal data type. Although they appear to be a number, they represent something similar to a name for the student and thus can't be quantitative (:Q) and also can't be ordinal (:O) since there isn't an order for names in this data set.
